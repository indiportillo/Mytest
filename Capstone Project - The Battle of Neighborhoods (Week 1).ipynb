##### Description of the problem

We will develop a program to help solve the problem of people seeking to relocate to King Couny, Washington. Although price is a very important variable, it is not the only one. For example, if the consumer is a family head, some other variables like the nearby will be important or the crime rate of that city. Those features can be obtained with the foursquare API

In short, we are going to carry out a program that helps consumers to solve what in Microeconomics is called "the consumer's problem", that is, consumers will try to maximize their utility taking into account a limited budget
